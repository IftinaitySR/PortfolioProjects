```python
# Safe extraction of title with error handling
title_element = soup2.find(id='productTitle')
if title_element:
    title = title_element.get_text().strip()
else:
    title = "Title not found"

# Safe extraction of price with multiple fallback options
price_element = soup2.find(id='priceblock_ourprice')
if price_element:
    price = price_element.get_text().strip()
else:
    # Try alternative price selectors as Amazon frequently changes these
    price_element = (soup2.find('span', class_='a-price-whole') or 
                    soup2.find('span', class_='a-offscreen') or
                    soup2.find('span', class_='a-price') or
                    soup2.find('span', {'class': 'a-price-range'}))
    if price_element:
        price = price_element.get_text().strip()
    else:
        price = "Price not found"

print(title)
print(price)
```

```python
# Create CSV and write headers and data into the file
import csv 
import datetime
import os

# Define the today variable (this was missing in your original code)
today = datetime.date.today()

# Make sure title and price variables exist (add default values if they don't)
try:
    title
except NameError:
    title = "Title not found"

try:
    price
except NameError:
    price = "Price not found"

header = ['Title', 'Price', 'Date']
data = [title, price, today]

# Get the current working directory to see where file will be saved
current_directory = os.getcwd()
print(f"File will be saved in: {current_directory}")

# Create the CSV file
try:
    with open('AmazonWebScraperDataset.csv', 'w', newline='', encoding='UTF8') as f:
        writer = csv.writer(f)
        writer.writerow(header)
        writer.writerow(data)
    print("CSV file created successfully!")
    print(f"File location: {os.path.join(current_directory, 'AmazonWebScraperDataset.csv')}")
except Exception as e:
    print(f"Error creating CSV file: {e}")
```
